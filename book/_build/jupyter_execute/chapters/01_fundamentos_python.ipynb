### Trabajo sobre datos.

Asumamos que, para un dato `x` y un dato `y` se obtuvieron los siguientes resultados.

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
# notar que x e y son listas de python, no arrays de numpy
```

Típicamente, uno no puede operar sobre listas de forma sencilla, por lo cual estos valores se transforman en `np.array` para operarlos como vectores (como los de nuestras clases de álgebra).

```python
x_vect = np.array(x)
y_vect = np.array(y)
```

Antes que todo, es importante obtener el tamaño de los vectores, para hacer un *sanity check*.

```python
print('tamaño de x:', x_vect.shape)
print('tamaño de y:', y_vect.shape)
```

Y ambos son vectores de largo 5, y no tienen otra dimensión. Por ahora no nos preocuparemos por el resto.

A continuación se verán operaciones básicas sobre estos vectores.  

```python
# suma de vectores
suma = x_vect + y_vect
print('suma de x e y:', suma)
# resta de vectores
resta = x_vect - y_vect
print('resta de x e y:', resta)
# producto de vectores elemento a elemento
producto = x_vect * y_vect 
# se multiplica cada elemento de x por el elemento correspondiente de y
print('producto de x e y:', producto)

# producto punto de x e y
producto_punto = np.dot(x_vect, y_vect)
print('producto punto de x e y:', producto_punto)
# tambien se puede usar @ para el producto punto
producto_punto_2 = x_vect @ y_vect
print('producto punto de x e y usando @:', producto_punto_2)
# son lo mismo 
```

Estas operaciones nos permitirán trabajar de forma rápida sobre los datos, especialmente cuando se necesitan generar descriptores de estos.
